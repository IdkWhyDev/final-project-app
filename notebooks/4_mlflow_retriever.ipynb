```python
import mlflow
from mlflow.sklearn import load_model
from mlflow.tracking import MlflowClient
from joblib import dump
```

```python
mlflow.set_tracking_uri("http://localhost:5000")
client = MlflowClient()

versions = client.search_model_versions("")
production_models = [m for m in versions if m.current_stage == "Production"]

for m in production_models:
    run = client.get_run(m.run_id)
    metrics = run.data.metrics
    print(m.name, metrics)
```

```python
mlflow.set_tracking_uri("http://localhost:5000")

model_name = "regmodel - tfidf_vectorizer - linear_svc"
model = load_model(f"models:/{model_name}/Production")
```

```python
dump(model, 'model.pkl')
```
