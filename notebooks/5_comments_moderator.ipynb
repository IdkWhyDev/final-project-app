```python
video_id = "wnGdTtj59zE"  # YouTube dummy

comments = []
next_page_token = None

while True:
    response = youtube.commentThreads().list(
        part="snippet",
        videoId=video_id,
        maxResults=100,
        pageToken=next_page_token
    ).execute()

    for item in response.get("items", []):
        comment_obj = item["snippet"]["topLevelComment"]
        comment_snippet = comment_obj["snippet"]

        comments.append({
            "id": comment_obj["id"],
            "author": comment_snippet["authorDisplayName"],
            "text": comment_snippet["textOriginal"]
        })

        # if "replies" in item:
        #     for reply in item["replies"]["comments"]:
        #         reply_snip = reply["snippet"]
        #         comments.append({
        #             "comment_id": reply["id"],
        #             "author": reply_snip.get("authorDisplayName"),
        #             "text": reply_snip.get("textDisplay")
        #         })

    next_page_token = response.get("nextPageToken")
    if not next_page_token:
        break

comments
```
